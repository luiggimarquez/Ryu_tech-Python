```python
import json
from json.decoder import JSONDecodeError
from pathlib import Path

data={}
path = Path('datos.txt')

if(path.is_file()):

	with open("datos.json", 'r') as f:
		try:
			result= f.read()
			data=json.loads(result)
		except JSONDecodeError:
			pass

def saveUser():

	while(True):

		nombre = input("Introduce nombre de usuario: ").lower()
		validation = data.get(nombre,'not found')
		if(validation == 'not found'):

			if((nombre != {}) and nombre.isalpha()):
				break
			else: print("Por favor introduce solo caracteres alfabéticos")

		else: print("Los usuarios deben ser unicos, ya hay un usuario con ese nombre")

	while(True):

		password = input(" Introduce password: ")
		if(password != ""):
			break
		else: print("Debes ingresar un password")

	data[nombre]=password

	with open("datos.json", 'w') as f:
		f.write( json.dumps(data, indent = 4) )
		f.close()


def consultUsers():

	print(" --- Usuarios registrados ---\n\n")
	if(data == {}):
		print("No hay usuarios registrados")
	else:
		print(data)

def login():
  
	print("-- login de usuario --")

	if(data != {}):
		login_user = input("Introduce tu usuario: ").lower()
		validation = data.get(login_user,'not found')

		if(validation == 'not found'):
			print("No se encuentra registrado")
		else:
			login_password = input("Introduce tu contraseña: ")

			if(validation == login_password):
				print("Login Successful")
			else: print("Login Failed")
	else:
		print("No hay usuarios registrados todavia!!!")

```

```python
saveUser()
```

```python
consultUsers()
```
